```python
import pandas as pd
df = pd.read_csv('fraud_detection.csv')
df.head(20)
```

```python
print("The number of transacion: ", len(df))
print(df[['type', 'amount']].head(20))
```

```python
df.drop_duplicates()
```

```python
#nu sunt valori dublicate
print("The number of transacion: ", len(df))
print(df[['type', 'amount']].head(20))
```

```python
#nr. de valori NaN
nan_count_total = df.isna().sum().sum()
print(nan_count_total)
```

```python
#inlocuire NaN cu 0
df_filled = df.fillna(0)
```

```python
nan2_count_total = df_filled.isna().sum().sum()
print(nan2_count_total)
```

```python
#procentajul de transactii frauduloase si nefrauduloase
print("Procent No Frauds :",df_filled['isFraud'].value_counts()[0]/len(df_filled['isFraud'])*100)
print("Procent Frauds :",df_filled['isFraud'].value_counts()[1]/len(df_filled['isFraud'])*100)
```

```python

columns_to_drop = ['isFlaggedFraud', 'nameOrig', 'nameDest']

# stergem coloanele
dff = df_filled.drop(columns=columns_to_drop, axis=1, inplace=False)
```

```python
print(dff.columns)
```

```python
#convertim variabila categoriala TYPE in 0 si 1
df_con = pd.get_dummies(data = dff,columns = ['type'], drop_first = True)
df_con.head()
```

```python
# Antrenare model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Split nonfraud and fraud data into train and test sets
nonfraud = df_con[df_con['isFraud']==0]
fraud = df_con[df_con['isFraud']==1]
nonfraud_train, nonfraud_test = train_test_split(nonfraud, test_size=0.2, random_state=42)
fraud_train, fraud_test = train_test_split(fraud, test_size=0.2, random_state=42)
```

```python
# Combinam seturile de antrenare
train_data = pd.concat([nonfraud_train, fraud_train])
```

```python
# Separare variabila features and target
features = train_data.drop('isFraud', axis=1)
target = train_data['isFraud']
```

```python
# Initializare model -- Logistic Regresion
model = LogisticRegression()

# Antrenare model
model.fit(features, target)

# Evaluare model pe setul de test
# Combinare seturi de test
test_data = pd.concat([nonfraud_test, fraud_test])
test_features = test_data.drop('isFraud', axis=1)
test_target = test_data['isFraud']

# Predictie folosind modelul antrenat
predictions = model.predict(test_features)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(test_target, predictions)
precision = precision_score(test_target, predictions)
recall = recall_score(test_target, predictions)
f1 = f1_score(test_target, predictions)
roc_auc = roc_auc_score(test_target, predictions)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print(f"ROC AUC Score: {roc_auc}")
```

```python
# Reprezentarea valorilor reale si cele prezise
results = pd.DataFrame({'Actual': test_target, 'Predicted': predictions})
print(results)
```
